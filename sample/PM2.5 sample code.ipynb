```python
with open('Songkhla.txt','r') as f:
    for i in range(12):
        print(f.readline())
```

```python
sk_pm25 = pd.read_csv('Songkhla.txt', sep='\t', header=None, skiprows=10)
sk_pm25
```

```python
sk_pm25['Time'] = pd.to_datetime({'year': sk_pm25[0], 'month': sk_pm25[1], 'day':sk_pm25[2],'hour': sk_pm25[3]})

sk_pm25['Time'] = sk_pm25['Time'].dt.tz_localize('UTC').dt.tz_convert('Asia/Bangkok')
sk_pm25['Time'] = sk_pm25['Time'].dt.tz_localize(None)

sk_pm25=sk_pm25.drop([0,1,2,3,5,6],axis=1)
sk_pm25.columns = ['PM2.5(µg/m3)', 'Time']
sk_pm25 = sk_pm25.set_index('Time')
```

```python
sk_pm25.head()
```

```python
fig = plt.figure(figsize=(28, 6))
plt.plot(sk_pm25,'rx')
```

```python
sk_pm25 = sk_pm25[~sk_pm25.index.duplicated(keep='first')]
sk_pm25.index = pd.DatetimeIndex(sk_pm25.index)
sk_pm25 = sk_pm25.reindex(pd.date_range("2016-03-03 15:00:00", "2020-03-18 20:00:00",freq='h'), fill_value=np.nan)
sk_pm25 = sk_pm25.interpolate()
```

```python
fig = plt.figure(figsize=(28, 6))
plt.plot(sk_pm25,'rx')
```
